```python
import pandas as pd
df = pd.read_csv('/home/yusuf/Masaüstü/Machine-learning/data/airplane_satisfied.csv')
```

```python
df.drop(['Unnamed: 0', 'id'], axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
flight_classes = ['Gender', 'Customer Type', 'Age', 'Type of Travel','Class']
for flight_class in flight_classes:
    print(df[flight_class].value_counts())
    sns.countplot(x=flight_class, data=df)
    plt.show()
```

```python
# Aykırı değerleri hesaplayalım
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Alt ve üst sınırlar
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerler
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Göster
print("Aykırı değer sayısı:", len(outliers))
display(outliers[['Age']].sort_values(by='Age'))
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df['Gender_encoded'] = df['Gender'].apply(lambda x: 1 if x=='Male' else 0)
df['Customer_Type'] = df['Customer Type'].apply(lambda x: 1 if x=='Loyal Customer' else 0)
df['Type_of_Travel'] = df['Type of Travel'].apply(lambda x: 1 if x=='Business travel' else 0)
from sklearn.preprocessing import LabelEncoder

#Elle belirlenmiş sıraya göre özel encoding yapmak için sınıfları yeniden ayarlayalım
class_mapping = {
    'Eco': 0,
    'Eco Plus': 1,
    'Business': 2
}
df['Class_encoded'] = df['Class'].map(class_mapping)
```

```python
df.Gender_encoded.value_counts()
```

```python
df[['Class', 'Class_encoded']].drop_duplicates().sort_values(by='Class_encoded')
```

```python
df.drop(['Gender','Class','Customer Type','Type of Travel'],axis = 1, inplace = True)
```

```python
df.columns
```

```python
print(df['Flight Distance'].value_counts())

sns.boxplot(x=df['Flight Distance'],data=df)
```

```python
#Flight distance aykırı değer temizleme
q1 = df['Flight Distance'].quantile(0.25)
q3 = df['Flight Distance'].quantile(0.75)
iqr = q3 - q1
lower = q1 - iqr*(1.5)
upper = q3 + iqr*(1.5)

df = df[(df['Flight Distance'] >= lower) & (df['Flight Distance'] <= upper)]
```

```python
df.shape
```

```python
df['Departure Delay in hours'] = df['Departure Delay in Minutes']/60
sns.boxplot(x=df['Departure Delay in hours'],data=df)
```

```python
q1 = df['Departure Delay in hours'].quantile(0.25)
q3 = df['Departure Delay in hours'].quantile(0.75)
iqr = q3 - q1
lower = q1 - iqr*(1.5)
upper = q3 + iqr*(1.5)

df_filtered = df[(df['Departure Delay in hours'] >= lower) & (df['Departure Delay in hours'] <= upper)]
```

```python
df_filtered.shape
```

```python
df_filtered['Arrival Delay in hours'] = df_filtered['Arrival Delay in Minutes']/60
sns.boxplot(x=df_filtered['Arrival Delay in hours'],data=df)
```

```python
q1 = df_filtered['Arrival Delay in hours'].quantile(0.25)
q3 = df_filtered['Arrival Delay in hours'].quantile(0.75)
iqr = q3 - q1
lower = q1 - iqr*(1.5)
upper = q3 + iqr*(1.5)

df_filtered = df_filtered[(df_filtered['Arrival Delay in hours'] >= lower) & (df_filtered['Arrival Delay in hours'] <= upper)]
```

```python
df_filtered.shape
```

```python
df_filtered.drop(['Departure Delay in Minutes','Arrival Delay in Minutes'], axis=1, inplace=True)
```

```python
df_filtered.columns
```

```python
col_features = ['Inflight wifi service','Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'satisfaction']


for col_feat in col_features:
    print(df_filtered[col_feat].value_counts())
```

```python
df_filtered['satisfaction'] = df_filtered['satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)
```

```python
df_filtered_copy = df_filtered.copy()
```

**dataset olusturma**

```python
from sklearn.preprocessing import StandardScaler

cols_to_scale = ['Age', 'Flight Distance', 'Departure Delay in hours', 'Arrival Delay in hours']

scaler = StandardScaler()
df_filtered[cols_to_scale] = scaler.fit_transform(df_filtered[cols_to_scale])
```

```python
df_filtered.head(3)
```

```python
X = df_filtered.drop(columns='satisfaction',axis=1)
y = df_filtered['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()
lr.fit(X_train, y_train)

accuracy_score(y_test,lr.predict(X_test))
```

```python
#tüm sütunların scale edilmiş hali ile
scaler = StandardScaler()
X_all = scaler.fit_transform(df_filtered_copy.drop(columns = 'satisfaction',axis = 1))
y_all = df_filtered_copy.satisfaction

X_all_train, X_all_test, y_all_train, y_all_test = train_test_split(X_all,y_all, random_state=42,test_size = 0.2)
lr = LogisticRegression()
lr.fit(X_all_train, y_all_train)

accuracy_score(y_all_test,lr.predict(X_all_test))
```
